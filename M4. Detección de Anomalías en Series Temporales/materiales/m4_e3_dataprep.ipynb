````python
# Importar librerías necesarias
import pandas as pd
from dataprep.eda import create_report

# Cargar un dataset de ejemplo (en este caso, utilizamos el dataset "tips" de seaborn)
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")

# Mostrar las primeras filas del dataset
print("Vista previa del dataset 'tips':")
display(df.head())
````

## Generación de Informe
La función `create_report` de Dataprep realiza un análisis exploratorio completo: describe las variables, detecta valores faltantes, genera gráficos interactivos, distribuciones y correlaciones, todo sin necesidad de codificar manualmente cada paso.

````python
# Generar un informe interactivo de EDA utilizando Dataprep
report = create_report(df)
````

## Visualización

Una vez generado, el informe se abre en el navegador, permitiendo explorar interactivamente los resultados y personalizar el análisis según se requiera.

````python
# Mostrar el informe en el navegador (para notebooks compatibles con Jupyter)
report.show_browser()  # También se puede usar report.show() en algunos entornos
````

Este ejercicio demuestra cómo, con pocas líneas de código, se puede obtener una visión completa del dataset gracias a las capacidades low-code de Dataprep.
